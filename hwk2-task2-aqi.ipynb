## About the data

In this task you will use [Air Quality Index (AQI)](https://www.airnow.gov/aqi/aqi-basics/) data from the [US Environmental Protection Agency](https://www.epa.gov) to visualize the impact on the AQI of the 2017 [Thomas Fire](https://en.wikipedia.org/wiki/Thomas_Fire) in Santa Barbara County. 

## FINAL CODE


```python
### ===== FINAL CODE =====
# load pandas using standard nomenclature
import pandas as pd

# read in 2017 by county data
aqi_17 = pd.read_csv('https://aqs.epa.gov/aqsweb/airdata/daily_aqi_by_county_2017.zip')

# read in 2018 by country data
aqi_18 = pd.read_csv('https://aqs.epa.gov/aqsweb/airdata/daily_aqi_by_county_2018.zip')

# (a) use head() to examine first 5 rows of data for 2017 
aqi_17.head()

# (b) use head() to examine first 5 rows of data for 2018
aqi_18.head()

# (c) look at names of columns in data frame for 2017
aqi_17.columns

# use concat() to combine dataframes from 2017 and 2018
aqi = pd.concat([aqi_17, aqi_18])

# rename columns to lower snake case
aqi.columns = aqi.columns.str.lower().str.replace(' ','_')

# select data from Santa Barbara County
aqi_sb = aqi[aqi.county_name=='Santa Barbara']

# check and make sure selection worked
aqi_sb

# remove state_name, county_name, state_code, and county_code
aqi_sb = aqi_sb.drop(columns=['state_name', 'county_name', 'state_code', 'county_code'])

# check the types of each column, date is listed as an object
aqi_sb.dtypes

# use to_datetime to chnge from object to date 
aqi_sb.date = pd.to_datetime(aqi_sb.date)

# check to make sure type was changed successfully
aqi_sb.dtypes

# update index to be the date column
aqi_sb = aqi_sb.set_index('date')

# check to see if index has been changed to date column, data type of the index is datetime64[ns]
aqi_sb.index

# created new column named five_day_average to aqi_sb that calculates the rolling 5 day mean
aqi_sb['five_day_average'] = aqi_sb.aqi.rolling('5D').mean()

# check to make sure the new column is present
aqi_sb.head()

# use plot function to create a line graph of aqi and 5 day average aqi
aqi_sb.plot(use_index=True,
            y=['aqi', 'five_day_average'],
            title='Air Quality Index in Santa Barbara County for 2017 & 2018',
            ylabel='Air Quality Index',
           color= ['pink', 'purple'])
```

Run the next cell and read through the comments. They will explain how to do make an average over a rolling window.

- We can createCreate a new variable `five_day_average` with the mean of the AQI over a 5-day rolling window. To do this use the `rolling`
- 

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.rolling.html


```python
# rolling() is a method for pandas.series that provides rolling window calculations
# the parameter '5D' indicates we want the window to be 5 days
# This is a lazy method (think groupby), we need to specify what we want to calculate over each window
# here we add the aggregator function mean()
# this indicates we want the mean over each window
# and we get a pd.Series as ouput
aqi_sb.aqi.rolling('5D').mean()
```
